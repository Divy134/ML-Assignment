## overlay two bar chart

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

a = np.array([2,8,6,4,5,7,3,1,2,5])
b = np.array([9,6,8,4,3,2,5,1,7,2])
ind = np.arange(len(a))

fig = plt.figure()
ax = fig.add_subplot(111)
ax.bar(x=ind, height=a, color = 'red', width=0.35,align='center')
ax.bar(x=ind, height=b, color = 'b', width=0.35/3,align='center')

plt.show()
```

```python
# create data
x = ['A', 'B', 'C', 'D']
x1 = [10, 40, 20, 30]
y2 = [25, 15, 35, 45]
 
# plot bars in stack manner
plt.bar(x, x1, color='y')
plt.bar(x, y2, bottom=x1, color='g')
plt.xlabel("x1")
plt.ylabel("y1")
plt.show()
```
